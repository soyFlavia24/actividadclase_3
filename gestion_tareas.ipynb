# crear tareas  
# Cada tarea debe ser representada como un diccionario con las claves nombre, descripcion y estado.

```python
tarea_1 = {"nombre" :"diseñar el modelo",
        "descripcion":["investigacion","planos","prototipos"],
        "estado":"terminado"},
tarea_2 ={"nombre":"seleccionar materiales",
          "descripcion":["eleccion de materiales","compras","presupuestos"],
          "estado":"en curso"},
tarea_3 ={"nombre":"elaboracion del mueble",
        "descripcion":["fabricacion","ensambles","manejo de herramientas"],
        "estado":"en curso"},
tarea_4 ={"nombre":"terminaciones",
          "descripcion":["evaluacion y ajustes","pruebas"],
          "estado":"pendiente"}
```
